## quesetion 1) How many people each employee manages?

```python
def countDirReport(bossId,df):
    count=0
    level=0
    reports=df.loc[df["boss_id"]==bossId,:].reset_index(drop =True)
    dirRptCnt=len(reports)
    count+=dirRptCnt
    #leafflag=False
    
    for i in range(dirRptCnt):
        report_i=reports.iloc[i,0]
        c_i=countDirReport(report_i,df)
        if c_i==0:
            #leafflag=True
            break
        else:
            count+=c_i 
            #level+=1
    
    return(count)
```

```python
cnt=len(hier)
hier["manage_cnt"]=hier["employee_id"].apply(countDirReport,df=hier)
```
