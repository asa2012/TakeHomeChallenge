```python
#create a table with day, employee_cnt, company id 
#define date range
unique_date=pd.date_range(start="2011-01-24",end="2015-12-13",freq="D")
unique_company=sorted(emp.company_id.unique())
```

```python
emp[emp['join_date']<="2015-3-16"]
```

```python
day=[]
company=[]
headcount=[]
for date in unique_date:
    for company_i in unique_company:
        num_hire=len(emp[(emp.join_date<=date) & (emp.company_id==company_i)])
        num_quit=len(emp[(emp.quit_date<=date) & (emp.company_id==company_i)])
        day.append(date)
        company.append(company_i)
        headcount.append(num_hire-num_quit)
        
```

```python
#create table as requested
table=pd.DataFrame({"date":day,"company":company,"headcount":headcount})
```

```python
# Choose the subset data
stop_date = pd.to_datetime('2015-12-13') - pd.DateOffset(days=365 + 31)
#data is up to end of 2015, therefore 2015 new hires data are not yet matured, we used one year ahead of max date, plus one extra month as cut off date  
#stop_date2 = pd.to_datetime('2015-12-13') - pd.DateOffset(months=13) #back to 13 months 
 
subset = emp[emp['join_date'] < stop_date]

#emp.count()
#subset.count()
# Binary label for early quit (less than 13 months)
quit = subset['quit_date'].isnull() | (subset['quit_date'] > subset['join_date'] + pd.DateOffset(days=396))
#quit is indexes for employees who did not quit, or emps who quit after one year tenure
subset['quit'] = 1 - quit.astype(int) #this 'quit' indicates emp quit within a year

subset.head()
```

```python
#method 1 to preprocessing features, decision tree only takes numeric value features 
subset['company_id']=subset['company_id'].astype(str)

dummies = pd.get_dummies(subset[['company_id', 'dept']]) 
#get_dummies only works for categorical only. If I did not do astype(str) above, get_dummies cannot convert it into indicator values
train_x=pd.concat(objs=[subset[['seniority','salary']],dummies],axis=1) 
#axis=1  ensures columns are concated 
train_y=subset['quit']
```

```python
train_x
train_y
```

```python
#method 2 to encode features
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
train_x=subset[['company_id','seniority','salary']]
train_x['dept']=le.fit_transform(subset['dept'])
train_y=subset['quit']
train_x.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf=DecisionTreeClassifier()
clf=clf.fit(X=train_x,y=train_y)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(clf, X=train_x,y=train_y, cv=10,scoring='f1_macro')
```

```python
features = list(train_x.columns)
target=['Not Quit','Quit'] 
features
```

```python
#print out decision tree into pdf file
from sklearn.externals.six import StringIO  
from sklearn import tree
import pydot 

dot_data2 = StringIO() 
tree.export_graphviz(clf, out_file=dot_data2,
                    feature_names = features,
                    class_names = target,
                    rounded = True, proportion = False, 
                    precision = 2, filled = True) 
graph2 = pydot.graph_from_dot_data(dot_data2.getvalue()) 
 
graph2[0].write_pdf("emp_retention.pdf") 
```
