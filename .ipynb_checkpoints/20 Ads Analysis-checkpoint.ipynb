# Question 1
If you had to identify the 5 best ad groups, which ones would they be? Which metric did you choose to identify the best ad groups? Why? Explain the pros of your metric as well as the possible cons. 

```python
import numpy as np
import pandas as pd
```

```python
ad_file='data/ad_table.csv'
ad=pd.read_csv(ad_file)
```

```python
ad.head()
```

```python
ad['shown'].sum()
```

```python
def calcMetrics(df):
    ctr=df['clicked'].sum()/df['shown'].sum()
    conv=df['converted'].sum()/df['shown'].sum()
    roi=df['total_revenue'].sum()/((df['clicked']*df['avg_cost_per_click']).sum())
    return pd.Series({'ctr':ctr,
                      'conv': conv,
                      'roi':roi})
```

```python
calcMetrics(ad)
```

```python
ad.groupby('ad').apply(calcMetrics).sort_values(by="roi",ascending=False)
```
