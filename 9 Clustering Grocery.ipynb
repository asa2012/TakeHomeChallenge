# question1, create user-item purchase matrix

```python
def id_to_item(df):
    """ function to convert id into counts """
    # 'sum' here is adding two lists into one big list
    ids = df['id'].str.split(',').sum()
    id_list = [0 for i in range(1, 49)]
    for i in ids:
        id_list[int(i) - 1] += 1
        
    return pd.Series(id_list, index=[*range(1,49)])
#[*range(1,49)] works in python 3,
#list(range(1,49)),works in python 2
```

```python
# get user_item count matrix
#purchase.head()
user_item_count = purchase.groupby(['user_id']).apply(id_to_item)
 
```

```python
user_count=user_item_count.sum(axis=1) #sum up by each user 
user_count.sort_values(ascending=False).head(5)
#user_item_count.sum(axis=0) #sum up by each item
```

## k-means
create feature matrix, rows and columns would be user and item
k-means clustering is a method of vector quantization, originally from signal processing, that is popular for cluster analysis in data mining. k-means clustering aims to partition n observations into k clusters in which each observation belongs to the cluster with the nearest mean, serving as a prototype of the cluster. This results in a partitioning of the data space into Voronoi cells.

```python
feature=user_item_count.T
feature.head()
#Here our goal is cluster different items, therefore matrix needs to be transposed, each row needs to be item 
```

```python
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score
```

## use kmeans to do cluster analysis 

```python
# determine the best number of clusters
clusters = range(2, 30)
inertias = [] #Sum of squared distances of samples to their closest cluster center.
#inertias lower the better 
silhouettes = []
#silhouettes higher the better, max value is 1
#The Silhouette Coefficient is calculated using the mean intra-cluster distance (a) and the mean nearest-cluster distance (b) for each sample. 
#The Silhouette Coefficient for a sample is (b - a) / max(a, b). 
#To clarify, b is the distance between a sample and the nearest cluster that the sample is not a part of. 
#Note that Silhouette Coefficient is only defined if number of labels is 2 <= n_labels <= n_samples - 1.

for n_clusters in clusters:
    kmeans = KMeans(n_clusters=n_clusters, init='k-means++', random_state=42, n_jobs=-1)
    kmeans = kmeans.fit(feature)
    label = kmeans.predict(feature)
    
    inertias.append(kmeans.inertia_)
    silhouettes.append(silhouette_score(feature, label))
```
